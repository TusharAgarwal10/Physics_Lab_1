```python
# credit: Shreyansh Kumar
import numpy as np
import matplotlib.pyplot as plt
```

```python
elength = np.array([6.7,13.5,19.3,26.1,32.0,37.7,42.9,47.6,51.9, 56.0, 59.5, 62.7, 65.4, 67.8, 69.8, 71.5, 72.2, 73.6, 74.3, 74.6])
len(elength)
```

```python
4.7/20
```

```python
z = []
for i in range(1, 21):
    olength = round(0.235*i, 1)
    z.append(olength)
print(z)
```

```python
average = np.polyfit(z, elength, 2)
print(average)

#check = 2*average[0]*np.array(z)+ average[1]

a = average[0]
b = average[1]
c = average[2]

#y = ax^2+bx+c
y = a * (np.array(z))**2 + b * np.array(z) + c
```

```python
plt.scatter(z, elength, color = "red", label = "Data Points")
plt.plot(z, y, label = "Fit")
plt.xlabel("Original Length (cm)")
plt.ylabel("Extended Length (cm)")
plt.legend()
plt.text(2.5, 40,fr'$y = {np.round(a,2)}x + {np.round(b,2)}y + ({np.round(c,2)}) $')
plt.show()  
```

```python
accurate_elength = ((180/1000)*9.81/1.2)*(((np.array(z)/100)/(4.7/100))-(1/2)*((np.array(z)/100)/(4.7/100))**2)

accurate_average = np.polyfit(z, accurate_elength, 2)
print(accurate_average)

#check = 2*average[0]*np.array(z)+ average[1]

p = accurate_average[0]
q = accurate_average[1]
r = accurate_average[2]

#y = ax^2+bx+c
y = p * (np.array(z))**2 + q * np.array(z) + r

plt.plot(z, y)
plt.scatter(z, accurate_elength, color = "red")
plt.show()
```
